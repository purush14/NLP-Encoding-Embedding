# Term Frequency - Inverse Document Frequency

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
documents = ["people watch movie and watch movie again", 
             "people watch cricket and watch cricket",
             "people like movie and like movie a lot",
             "people like cricket"]
```

```python
tf_idf = TfidfVectorizer()
```

```python
tf_idf_vector = tf_idf.fit_transform(documents)
```

```python
print("TF-IDF Vector Shape:", tf_idf_vector.shape)
print("TF-IDF Vocabulary:", tf_idf.get_feature_names_out())
print("TF-IDF Vectors (dense format):\n", tf_idf_vector.todense())
print("TF-IDF Vectors:\n", tf_idf_vector.toarray())
```

```python
for vector in tf_idf_vector.toarray():
    print(vector)
```
